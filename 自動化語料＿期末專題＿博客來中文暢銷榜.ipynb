### 找出博客來中文暢銷榜前100名
書名/作者/價格/購買連結/封面圖片連結

```python
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import os

def get_soup(url, post_data=None):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    }

    try:
        resp = requests.get(url, headers=headers)
        resp.encoding = "utf-8"
        if resp.status_code == 200:
            soup = BeautifulSoup(resp.text, "lxml")
            return soup
        else:
            print("網頁取得失敗", resp.status_code)
    except Exception as e:
        print("網址錯誤", e)
```

```python
resp = get_soup("https://www.books.com.tw/web/sys_saletopb/books/?loc=P_menu_th_1_002")
resp
```

```python
soup = resp
```

```python
soup.find(class_ = "mod_a clearfix")
```

```python
books = soup.find(class_ = "mod_a clearfix").find_all("li", class_ = "item")
books
```

```python
book = books[0]
```

```python
#書名
book.find("h4").text
```

```python
#作者
book.find("li").text.split("：")[1].strip()
```

```python
#價格
book.find(class_ = "price_a").text.split("折")[1].strip()
```

```python
#購買連結
book.find("a").get("href")
```

```python
#封面圖片連結
book.find(class_ = "cover").get("src")
```

```python
data = []

for book in books:
    try:
        title = book.find("h4").text
        author = book.find("li").text.split("：")[1].strip()

        # 檢查 "折" 是否存在
        price_text = book.find(class_="price_a").text
        if "折" in price_text:
            price = price_text.split("折")[1].strip()
        else:
            price = None

        link = book.find("a").get("href")
        cover = book.find(class_="cover").get("src")
        data.append([title, author, price, link, cover])
    except Exception as e:
        print(e)

bb = pd.DataFrame(data, columns=["書名", "作者", "價格", "購買連結", "封面圖片連結"])
bb
```

```python
bb.to_csv("博客來中文暢銷榜.csv", index=False)
```
