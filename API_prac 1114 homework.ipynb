### API

- 與網頁互動，網址會切換，且中間有「?」表示有API居中以get的方式處理資料傳輸。「?」後面即跟網站要資料的參數資料。
- 與網頁互動，網址都不變，表示有API居中以POST的方式處理資料傳輸。

#### API post的例子：ibon網站
- https://www.ibon.com.tw/retail_inquiry.aspx#gsc.tab=0

```python
# 用一般方式抓回的資料，也沒有網頁查詢的結果

import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import os

def get_soup(url, post_data=None):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    }

    try:
        resp = requests.get(url, headers=headers)
        resp.encoding = "utf-8"
        if resp.status_code == 200:
            soup = BeautifulSoup(resp.text, "lxml")
            return soup
        else:
            print("網頁取得失敗", resp.status_code)
    except Exception as e:
        print("網址錯誤", e)
resp=get_soup('https://www.ibon.com.tw/retail_inquiry.aspx#gsc.tab=0')
resp
```

```python
# 由request method=post的header找到網址

api_url='https://www.ibon.com.tw/retail_inquiry_ajax.aspx'
```

```python
# 組合FormData（互動的參數）
# 在payload頁籤裡

form_data = {'strTargetField': 'COUNTY',
             'strKeyWords': '基隆市'}

form_data
```

```python
# requests method不是get，要用post

import requests

resp=requests.post(api_url, form_data)
```

```python
# 確定OK後，再用BeautifulSoup來做資料抽取

from bs4 import BeautifulSoup
```

```python
# resp要重新指定給requests.post，不然會抓到之前requests.get那個

soup=BeautifulSoup(resp.text, 'lxml')
soup
```

```python
trs=soup.find('table', class_='font16').find_all('tr')
```

```python
trs
```

```python
tds = trs[1].find_all('td')
tds
```

```python
for td in tds:
    shop=td.text.strip()
    print(shop)
```

```python
for td in tds:
    shop = td.find()
    print(shop)
```

```python
datas = []
for tr in trs[1:]: #假設trs是一個大串列，從第二個[1]開始來跳過標題
    tds = tr.find_all("td")
    #因為網站已經把資料整理好了，所以按照它的排法，店號在第一排、店名在第二排、地址在第三排
    store_number = tds[0].text.strip()
    store_name = tds[1].text.strip()
    address = tds[2].text.strip()
    #把資料加進datas中
    datas.append([store_number, store_name, address])
datas
```

```python
df = pd.DataFrame(datas,columns = ["店號", "店名", "地址"])
df
```
